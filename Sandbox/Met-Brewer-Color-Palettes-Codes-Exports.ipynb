```python
# Version 4
import met_brewer
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Print the colors with their codes for each palette
def printPalette(palette_name, hex_colors):

    # Convert hex colors to RGB tuples for imshow
    rgb_colors = [mcolors.hex2color(color) for color in hex_colors]

    # Convert from floats (0-1) to integers (0-255)
    rgb_ints = [str(tuple(int(component * 255) for component in color)) for color in rgb_colors]

    print("Color Codes: ")
    for idx, hex_color in enumerate(hex_colors):
        print(idx + 1, ". HEX: ", hex_color, ",\tRGB: " ,rgb_ints[idx])
    print("\n")

    ## Previous Color Codes print
    # print("RGB: \n", ",".join([str(tuple(int(component * 255) for component in color)) for color in rgb_colors]).replace(" ", ""))
    # print("HEX: \n", ", ".join(hex_colors))


    # Set up the figure
    plt.figure(figsize=(8, 3), dpi=150)

    # Display the colors (using rgb_colors ensures order matches hex_colors)
    plt.imshow([rgb_colors], aspect='auto')

    # Add hex labels for each color
    for idx, hex_color in enumerate(hex_colors):

        # Determine text color for readability based on color brightness
        rgb = rgb_colors[idx]

        # Simple brightness calculation for RGB in 0-1 range
        text_color = 'black' if (rgb[0]*0.299 + rgb[1]*0.587 + rgb[2]*0.114) > 0.5 else 'white'
        plt.text(idx, 0, idx + 1, ha='center', va='bottom', color=text_color, fontsize=15)
        # # Previous labels (Instead of numbers, display the hex codes)
        # plt.text(idx, 0, hex_color, ha='center', va='bottom', color=text_color, fontsize=10, rotation=45)


    # Set the title with colorblind-friendly check
    title_suffix = " (Colorblind Friendly)" if met_brewer.is_colorblind_friendly(palette_name) else ""
    plt.title(palette_name + title_suffix)

    plt.axis('off')  # Hide the axis for a cleaner look
    plt.show()
    plt.close()







# Loop through all palettes
for palette_name in met_brewer.MET_PALETTES.keys():
    print("\n\n\n")

    # Get the palette and description
    palette = met_brewer.met_brew(palette_name)

    # Get the hex colors list directly
    hex_colors = list(met_brewer.export(palette_name).values())[1]

    # Print the palette
    printPalette(palette_name, hex_colors) # if this line is number is 78 => exactly 2 pages.

    # Align the next palette to the next page
    lines_per_pages = 42 # Approx 43 lines per page.
    lines = 8 + len(hex_colors) + 12 # 5(=2 empty line begging + 2 Headings), 12 for the colors plot (approx)
    for i in range(lines_per_pages - lines):
        print()
```
